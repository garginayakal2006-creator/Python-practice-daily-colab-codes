```python

#phy+coding
def pressure (force, area) :
  P=force/area
  print("pressure is =", P, "pascal")

  if P>50 :
    print("Ahh! It seems pressure is high.are you stressed😖")
  else :
    print(" Yepp! pressure is low, are you happy😁? ")

force=500
area=5
pressure (force, area)
```

```python
def velocity (distance, time) :
  v=distance/time
  print ("velocity =", v, "m/s")

  if v>50:
   print("you are to fast🚴, slow down dude! ")
  elif  50< v>20 :
   print("Great speed! It seems you are following traffic rules  😅")
  else :
   print("Hmm... To slow🐌,need redbull I think😂")
  if v%5==0:
   print(" Good speed")

distance =70
time=2
velocity (distance, time)
```

```python
def function(force, distance, time):
  work=force*distance
  power=work/time
  print("work =",work, "joules")
  print ("power=", power, " watt")

  if power>=100:
    print ("Incredible power! ⚡ You’re unstoppable!")
  elif (100>power>50) :
    print ("Good power output 💪 Keep going!")
  else:
    print("Low power 🐢 Maybe time for a recharge?")


force= 50
distance=10
time=5
function (force, distance, time)
```
